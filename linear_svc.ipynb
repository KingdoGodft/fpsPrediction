```python
import numpy as np
from sklearn.linear_model import Lasso
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import asgl


frame_data = pd.read_excel('FPS_DATA.xlsx')

frame_data.head()
```

```python
frame_Y = pd.DataFrame(frame_data['GAME AVG FRAME'], columns = ['GAME AVG FRAME'])
frame_Y[:5]
```

```python
scale_cols = ['CPU BenchMark', 'GPU BenchMark', 'CPU Core', 'CPU Thread', 'GPU Memory GB', 'GPU Clock Boost MHz', 'GAME NAME']
data_temp  = frame_data[scale_cols]
frame_X = pd.DataFrame(data_temp)
```

```python
frame_df = pd.concat([frame_Y, frame_X], axis=1)
```

```python
def lasso_regress(data, yvar, xvars):
    # output, input variables

    Y = data[yvar]
    X = data[xvars]

    # Create lasso train and test model

    X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)


    # Fit the lasso model

    lasso = Lasso().fit(X_train, y_train)
   
    # Get the intercept and coefficients

    print("훈련 세트의 정확도 : {:.2f}".format(lasso.score(X_train, y_train)))
    print("테스트 세트의 정확도 : {:.2f}".format(lasso.score(X_test, y_test)))
    print("사용한 특성의 수 : {}".format(np.sum(lasso.coef_ != 0)))
    print("사용한 max_iter : {}".format(lasso.n_iter_))


  

    return 0
```

```python
# Groupby
grouped = frame_df.groupby('GAME NAME')

#Apply Lasso by Group
lasso_reg_coef = grouped.apply(lasso_regress, 'GAME AVG FRAME', ['CPU BenchMark', 'GPU BenchMark', 'CPU Core', 'CPU Thread', 'GPU Memory GB', 'GPU Clock Boost MHz'])
```
