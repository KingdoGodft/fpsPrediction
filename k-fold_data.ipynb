```python
# 모델 선언
fold_df_clf = DecisionTreeClassifier()

# Kfold
SPLITS = 5
kf = KFold(n_splits = SPLITS)
n_iter = 0
```

```python
label  = pd.DataFrame(frame_data['GAME NAME'])

features = frame_data

df_specs = pd.DataFrame(data=None, columns=frame_data.columns, index=frame_data.index)
df_specs.dropna()
```

```python
fold = 1
for train_index, validate_index  in kf.split(frame_data):
    trainDF = pd.DataFrame(frame_data.iloc[train_index])
    validateDF = pd.DataFrame(frame_data.iloc[validate_index])
    trainDF[['set', 'iteration']] = 'T', fold
    validateDF[['set', 'iteration']] = 'V', fold
    print("Fold #{}, Training Size: {}, Validation Size: {}".format(fold,len(trainDF),len(validateDF)))
    fold+=1
    df_specs = pd.concat([df_specs, validateDF,trainDF])

df_specs.to_csv("./auto-mpg-kfold5.csv")
print(df_specs)
```
